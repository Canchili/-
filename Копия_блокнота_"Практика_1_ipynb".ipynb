**Постройте корреляционную матрицу для всех числовых переменных и визуализируйте её с помощью тепловой карты:**

```python
df = df.drop('Gender', axis=1)
df = df.drop('Profession', axis=1)
# Корреляционная матрица
correlation_matrix = df.corr()

# Визуализация тепловой карты
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Корреляционная матрица')
plt.show()
```

Переменные Gender и Profession не коррелируются

сильную положительную корреляцию не имеет ни одна переменная, самая высокая это 0.33

сильную слабую корреляцию так же ни одна переменная не имеет, самая низкая это -0.042
